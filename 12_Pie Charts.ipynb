```python
import matplotlib.pyplot as plt
import numpy as np
```

# Pie Chart

```python
p = np.array([50,25,15,10])

plt.pie(p)
plt.show()
```

# Labels

```python
p = np.array([50,25,15,10])

mylabel = ["Mango","Apple", "Banana", "Orange"]


plt.pie(p, labels=mylabel)
plt.show()
```

# Angles

```python
p = np.array([50,25,15,10])

mylabel = ["Mango","Apple", "Banana", "Orange"]


plt.pie(p, labels=mylabel, startangle = 90)
plt.show()
```

# Explode

```python
p = np.array([50,25,15,10])

mylabel = ["Mango","Apple", "Banana", "Orange"]

myexplode = [0.1,0.1,0.1,0.1]

plt.pie(p, labels=mylabel, explode= myexplode)
plt.show()
```

# Shadow

```python
p = np.array([50,25,15,10])

mylabel = ["Mango","Apple", "Banana", "Orange"]



plt.pie(p, labels=mylabel, shadow=True)
plt.show()
```

# colors

```python
p = np.array([50,25,15,10])

mylabel = ["Mango","Apple", "Banana", "Orange"]

mycolor = ['yellow', 'red', 'green', 'orange']

plt.pie(p, labels=mylabel, colors=mycolor)
plt.legend()
plt.show()
```

# Legend

```python
p = np.array([50,25,15,10])

mylabel = ["Mango","Apple", "Banana", "Orange"]



plt.pie(p, labels=mylabel,)
plt.legend()
plt.show()
```

# Legend with header

```python
p = np.array([50,25,15,10])

mylabel = ["Mango","Apple", "Banana", "Orange"]



plt.pie(p, labels=mylabel)
plt.legend(title= '4 Fruits')
plt.show()
```
