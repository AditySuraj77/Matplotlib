```python
import matplotlib.pyplot as plt
import numpy as np
```

# Grid

```python
x = np.array([10,20,30,40,50,60,70,80,90])
y = np.array([5,10,15,20,25,30,35,40,45])

plt.title('None', c= 'red', font = 'serif', )
plt.xlabel('Null')
plt.ylabel('Null')
plt.plot(x,y)
plt.grid()
plt.show()
```
