# Format Strings

```python
plt.plot(ypoint, 'o:k')
plt.show()
```

```python
plt.plot(ypoint, marker = 'o' , ms='20')
plt.show()
```

```python
plt.plot(ypoint, marker = 'o' , ms='20', mec="r" )
plt.show()
```

```python
plt.plot(ypoint, marker = 'o' , ms='20', mec="y", mfc='y' )
plt.show()
```

```python
plt.plot(ypoint, marker = 'o' , ms='20', mec="hotpink", mfc='hotpink' )
plt.show()
```
