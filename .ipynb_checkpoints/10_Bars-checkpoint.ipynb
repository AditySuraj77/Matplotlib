```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(['A','B','C','D','C','E'])
y = np.array([1,2,3,2,1,1])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(['A','B','C','D','C','E'])
y = np.array([1,2,3,2,1,1])

plt.barh(x,y)
plt.show()
```

```python
x = np.array(['A','B','C','D','C','E'])
y = np.array([1,2,3,2,1,1])

plt.bar(x,y, color='y')
plt.show()
```

```python
x = np.array(['A','B','C','D','C','E'])
y = np.array([1,2,3,2,1,1])

plt.barh(x,y, color='y')
plt.show()
```

```python
x = np.array(['A','B','C','D','C','E'])
y = np.array([1,2,3,2,1,1])

plt.bar(x,y, color='green', width = 0.2)
plt.show()
```

```python
x = np.array(['A','B','C','D','C','E'])
y = np.array([1,2,3,2,1,1])

plt.barh(x,y, color='aqua', height = 0.3)
plt.show()
```
