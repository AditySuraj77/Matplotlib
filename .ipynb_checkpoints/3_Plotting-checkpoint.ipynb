```python
import matplotlib.pyplot as plt
import numpy as np
```

## Plotting without line

```python
x = np.array([10])
y = np.array([10])
plt.plot(x,y , "o")
plt.show()
```

## Multiple Points

```python
xpoint = np.array([5,10,15,20,25])
ypoint = np.array([0,200,30,300,50])

plt.plot(xpoint, ypoint)
plt.show()
```

## Default X-Points

```python
ypoint = np.array([3,8,1,5,10])

plt.plot(ypoint)
plt.show()
```
