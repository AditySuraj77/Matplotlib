```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

```python
plt.plot(ypoints, ls= '--')
plt.show()
```

```python
plt.plot(ypoints, ls='-.')
plt.show()
```

# Line color

```python
plt.plot(ypoints, c='g')
plt.show()
```

```python
plt.plot(ypoints, color='hotpink')
plt.show()
```

```python
plt.plot(ypoints, c='hotpink', lw= '15')
plt.show()
```

```python
plt.plot(ypoints, c='g', linewidth='10')
plt.show()
```

# Multiple Line

```python
y1 = np.array([1,8,3,12])
y2 = np.array([2,10,5,10])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```
