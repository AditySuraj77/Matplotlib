```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,4,6,7,8,6,9,8,7,5,6,4,4])
y = np.array([5,9,8,7,7,5,4,5,3,5,3,2,5,5])

plt.scatter(x,y)
plt.show()
```

```python
x = np.array([1,2,40,6,7,80,6,9,80,7,5,60,4,4])
y = np.array([5,90,8,70,7,5,40,5,3,50,3,2,50,5])

plt.scatter(x,y, c= 'r')




x = np.array([10,20,4,60,70,8,6,90,8,7,5,60,4,40])
y = np.array([5,90,8,70,7,50,4,5,3,50,3,2,50,5])

plt.scatter(x,y, c= 'b')
plt.show()
```

```python
x = np.array([10,20,4,60,70,8,6,90,8,7,5,60,4,40])
y = np.array([5,90,8,70,7,50,4,5,3,50,3,2,50,5])

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta","Aqua"])



plt.scatter(x,y, c= colors )

plt.show()
```

```python
x = np.array([10,20,4,60,70,8,6,90,8,7,5,60,4,40])
y = np.array([5,90,8,70,7,50,4,5,3,50,3,2,50,5])

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta","Aqua"])



plt.scatter(x,y, c= colors, cmap="viridis_r" )
plt.colorbar()
plt.show()
```

```python
x = np.array([1,2,40,6,7,80,6,9,80,7,5,60,4,4])
y = np.array([5,90,8,70,7,5,40,5,3,50,3,2,50,5])

plt.scatter(x,y, c= 'r')




x = np.array([10,20,4,60,70,8,6,90,8,7,5,60,4,40])
y = np.array([5,90,8,70,7,50,4,5,3,50,3,2,50,5])

sizes = np.array([1000,20,30,40,50,600,70,80,90,100,110,120,130,140])
plt.scatter(x,y, c= 'b', s=sizes, alpha=0.5)
plt.show()
```
